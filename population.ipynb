```python
import math
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace, zeros, exp

def logistic (t):
        return 5000000./(1+3000000*math.exp(-1*t))

x = np.linspace(0,30,31)
y = [logistic(i) for i in x]
plt.scatter(x,y)
plt.plot(x, y)
plt.show()
```

```python
def ode_FE(f, U_0, dt, T):
    N_t = int(round(float(T)/dt))
    u = zeros(N_t+1)
    t = linspace(0, N_t*dt, len(u))
    u[0] = U_0
    for n in range(N_t):
        u[n+1] = u[n] + dt*f(u[n], t[n])
    return u, t
```

```python
#Birth rate	2.97 births/1,000 population (2014 est.)
#Death rate	1.76 deaths/1,000 population (2014 est.)
br = 2.97/1000
dr = 1.76/1000

rm = br-dr
rex = 0.15

dt = 1
T =30 

u, t = ode_FE(f=lambda u, t: 3.7/1000*(1 - u/5000000.)*u, \
                           U_0=3000000, dt=dt, T=T)
plt.figure()  # Make separate figures for each pass in the loop
plt.plot(t, u, 'b-')
plt.xlabel('t'); plt.ylabel('N(t)')
plt.savefig('tmp_%g.png' % dt); plt.savefig('tmp_%g.pdf' % dt)

u, t = ode_FE(f=lambda u, t: 30/1000*(1 - u/5000000.)*u, \
                           U_0=3000000, dt=dt, T=T)
plt.figure()  # Make separate figures for each pass in the loop
plt.plot(t, u, 'b-')
plt.xlabel('t'); plt.ylabel('N(t)')
plt.savefig('tmp_%g.png' % dt); plt.savefig('tmp_%g.pdf' % dt)
```
